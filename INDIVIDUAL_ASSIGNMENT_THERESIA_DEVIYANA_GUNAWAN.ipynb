##Plot the following in Python Code

####Unit step function u(t) in the interval -5 < t < 5

```python
import numpy as np
import matplotlib.pyplot as plt
import scipy as signal
```

```python
def u(t):
    return 1 * (t >= 0)
t1 = np.linspace(-5, 5, 100)
plt.plot(t1,u(t1),'g')
plt.title("UNIT STEP FUNCTION -5<t<5");
```

####Unit step function u(t) delayed 2 units

```python
def u(t):
    return 1 * (t >= 0)
t1 = np.linspace(-6, 10, 1000)
delayed = u(t1-2)
plt.plot(t1,delayed,'r')
plt.title("UNIT STEP FUNCTION DELAYED 2 UNITS");
```

###u(t) function advanced by 2 units

```python
def u(t):
    return 1 * (t >= 0)
t1 = np.linspace(-6, 10, 1000)
advanced = u(t1+2)
plt.plot(t1,advanced,'m')
plt.title("UNIT STEP FUNCTION ADVANCED 2 UNITS");
```

###Impulse signal -5<t<5

```python
def x(t):
    return 1 * (t == 0)
t1 = np.linspace(-5, 5, 13)
plt.plot(t1,x(t1),'m')
plt.title("Unit impulse function δ(t) in the interval -5 < t < 5");
```

###δ(t) delayed by 4 units

```python
def x(t):
    return 1 * (t == 0)
t1 = np.linspace(-8, 8, 13)
xsignal1 = x(t1-4)
plt.plot(t1,xsignal1)
plt.title("Unit impulse function δ(t) delayed by 4 units");
```

###δ(t) advanced by 2 units

```python
def x(t):
    return 1 * (t == 0)
t1 = np.linspace(-6, 6, 13)
xsignal1 = x(t1+2)
plt.plot(t1,xsignal1)
plt.title("Unit impulse function δ(t) advanced by 2 units");
```

###y(t)=3u(t+2)-u(t-4)

```python
def u(t):
    return 1 * (t >= 0)
t1 = np.linspace(-5, 10, 1000)
y = 3*u(t1+2) - u(t1-4)
plt.plot(t1,y,'g')
plt.title("y(t)=3u(t+2)-u(t-4)");
```

##Sampling Theorem
* A baseband signal x(t), band limited to 2πβ radians/s, can be 
reconstructed accurately from its samples x(kT), if the sampling 
rate ω0, in radians/s, satisfies filtering condition:
ω0 ≥ 4πβ
In terms of sampling rate 
f0 = ω0/2π in samples/s 
or in terms of sampling intervals Ts
To prevent errors (aliasing), sampling rate (samples/s) must be
f0 ≥ 2β
Or
Sampling interval
T_0 ≤
1/2β

* frequency 1 Hz
* sampling rate 10 Hz (sampling theorem)

```python
t = np.linspace(0,1,500)
w = 1
x = np.sin(2*np.pi*w*t)

sampling_rate = 10 # the larger the sampling rate, the better the resulting graph f0 ≥ 2β
T = 1/sampling_rate
n = np.arange(0,1/T)
nT = n*T
x2 = np.sin(2*np.pi*w*nT)

plt.figure(figsize=(10,8))
plt.subplot(2,2,1)
plt.plot(t,x,label = "Sine Wave Frequency 1 Hz")
plt.legend(loc="upper right")

plt.subplot(2,2,2)
plt.plot(nT,x2,'ro',label="Sample with sample rate 10 Hz")
plt.legend(loc="upper right")

plt.subplot(2,2,3)
plt.stem(nT,x2,label="Sample with sample rate 10 Hz")
plt.legend(loc="upper right")

plt.subplot(2,2,4)
plt.plot(nT,x2,'m',label="reconstruction signal after sampling")
plt.legend(loc="upper right")
plt.tight_layout()
;
```

```python
#Plots

plt.subplots(1,3,figsize=(15,5))

#x[n]
plt.subplot(131)
plt.scatter(t_x,x_n)
plt.ylim(-3, 3)
plt.xlim(0,9)
plt.title('x[n]')
plt.xlabel('Sample no.')
plt.grid()

#h[n]
plt.subplot(132)
plt.scatter(t_h,h_n)
plt.ylim(-3,3)
plt.xlim(0,4)
plt.title('h[n]')
plt.xlabel('Sample no.')
plt.grid()

#Convolution
conv = signal.convolve(x_n,h_n)
t = np.linspace(0, len(conv)-1, len(conv))
plt.subplot(1,3,3)
plt.plot(t,conv,'or')
plt.xlim(0,13)
plt.ylim(-3,3)
plt.grid()
plt.title('x[n]*h[n]')
plt.xlabel('Sample no.');
#plt.tight_layout()
```

### Perform the convolution of following signal pairs 

* Sine wave: 1 Hz, Sine wave: 5 Hz

```python
#Sine wave 1 Hz
t = np.linspace(0,1,1000)
w = 1
x1 = np.sin(2*np.pi*w*t)

#Sine wave 5 Hz
t = np.linspace(0,1,1000)
w = 5
x2 = np.sin(2*np.pi*w*t)

#plot

plt.subplots(1,3,figsize=(15,5))

#sine wave 1 Hz
plt.subplot(131)
plt.plot(t,x1,'r')
plt.title('x(t)')
plt.grid()

#sine wave 5 Hz
plt.subplot(132)
plt.plot(t,x2)
plt.title('h(t)')
plt.grid()

#Convolution
conv = signal.convolve(x1,x2)
t = np.linspace(0, len(conv)-1, len(conv))
plt.subplot(1,3,3)
plt.plot(t,conv,'m')
plt.grid()
plt.title('x(t)*h(t)')
#plt.tight_layout()
```

* Sine wave, 2Hz, Cosine wave: 5Hz 

```python
#Sine wave 2 Hz
t = np.linspace(0,1,1000)
w = 2
x1 = np.sin(2*np.pi*w*t)

#cosine wave 5 Hz
t = np.linspace(0,1,1000)
w = 5
x2 = np.cos(2*np.pi*w*t)

#plot

plt.subplots(1,3,figsize=(15,5))

#sine wave 2 Hz
plt.subplot(131)
plt.plot(t,x1,'r')
plt.title('x(t)')
plt.grid()

#cosine wave 5 Hz
plt.subplot(132)
plt.plot(t,x2)
plt.title('h(t)')
plt.grid()

#Convolution
conv = signal.convolve(x1,x2)
t = np.linspace(0, len(conv)-1, len(conv))
plt.subplot(1,3,3)
plt.plot(t,conv,'m')
plt.grid()
plt.title('x(t)*h(t)')
#plt.tight_layout()
```

* Cosine wave: 2Hz, Cosine wave: 2Hz + phase diff: $pi$ 

```python
#cosine wave 2 Hz
t = np.linspace(0,1,1000)
w = 2
x1 = np.cos(2*np.pi*w*t)

#cosine wave 2 Hz + pi
t = np.linspace(0,1,1000)
w = 2
x2 = np.cos(2*np.pi*w*t + np.pi)

#plot

plt.subplots(1,3,figsize=(15,5))

#cosine wave 2 Hz
plt.subplot(131)
plt.plot(t,x1,'r')
plt.title('x(t)')
plt.grid()

#cosine wave 2 Hz + pi
plt.subplot(132)
plt.plot(t,x2)
plt.title('h(t)')
plt.grid()

#Convolution
conv = signal.convolve(x1,x2)
t = np.linspace(0, len(conv)-1, len(conv))
plt.subplot(1,3,3)
plt.plot(t,conv,'m')
plt.grid()
plt.title('x(t)*h(t)')
#plt.tight_layout()
```
